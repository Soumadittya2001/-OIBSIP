```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

from google.colab import files
uploaded = files.upload()

import io
df = pd.read_csv(io.BytesIO(next(iter(uploaded.values()))))


pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

def display_scrollable(dataframe):
    display(HTML(dataframe.to_html().replace(
        '<table border="1" class="dataframe">',
        '<div style="overflow-x:auto; max-width:100%"><table border="1" class="dataframe">'
    ) + '</div>'))

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print("📌 First 5 Rows:")
display_scrollable(df.head())

print("\n📌 Data Info:")
print(df.info())

print("\n📌 Missing Values:")
print(df.isnull().sum())


df.fillna("Unknown", inplace=True)

print("\n📊 Descriptive Statistics:")
display_scrollable(df.describe())

print("\n📊 Median:")
print(df.median(numeric_only=True))

print("\n📊 Mode:")
display_scrollable(df.mode().iloc[[0]])

print("\n📊 Standard Deviation:")
print(df.std(numeric_only=True))


if 'category' in df.columns:
    print("\n🗂️ Unique Categories:")
    print(df['category'].value_counts())


if 'item' in df.columns and 'calories' in df.columns:
    top_items = df[['item', 'calories']].sort_values(by='calories', ascending=False).head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(data=top_items, x='item', y='calories')
    plt.title("Top 10 High-Calorie Items")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

if 'category' in df.columns and 'calories' in df.columns:
    avg_cat = df.groupby('category')['calories'].mean().sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    avg_cat.plot(kind='bar', color='orange')
    plt.title("Average Calories per Category")
    plt.ylabel("Average Calories")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

plt.figure(figsize=(10, 6))


numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()
corr_matrix = corr_matrix[(corr_matrix.abs() >= 0.3) | (corr_matrix == 1.0)]

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    annot_kws={"size": 10}
)

plt.title("Filtered Correlation Heatmap", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()


```
